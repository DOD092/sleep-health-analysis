```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/preprocess/eda_pre.csv", header=None)
df.head()
```

```python
columns = [
    "Person_ID",
    "Gender",
    "Age",
    "Occupation",
    "Sleep_Duration",
    "Quality_of_Sleep",
    "Physical_Activity_Level",
    "Stress_Level",
    "BMI_Category",
    "Blood_Pressure",
    "Heart_Rate",
    "Daily_Steps",
    "Sleep_Disorder",
    "Systolic",
    "Diastolic",
    "Blood_Pressure_Category"
]

df.columns = columns

df.head()
```

```python
df["Sleep_Duration"] = df["Sleep_Duration"].astype(str).str.replace(",", ".")
df["Sleep_Duration"] = df["Sleep_Duration"].astype(float)
```

```python
df.info()
```

```python
num_cols = [
    "Age",
    "Quality_of_Sleep",
    "Physical_Activity_Level",
    "Stress_Level",
    "Heart_Rate",
    "Daily_Steps",
    "Systolic",
    "Diastolic"
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
df.isnull().sum()
```

```python
df["Sleep_Disorder"] = df["Sleep_Disorder"].fillna("None")
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Person_ID", "Blood_Pressure","Diastolic","Blood_Pressure_Category"])
```

```python
text_cols = ["Gender", "Occupation", "BMI_Category", "Sleep_Disorder"]

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
df.head(5)
```

```python
df.to_csv("../dataset/preprocess/sleep_health_clean.csv", index=False)

print("Done")
```
